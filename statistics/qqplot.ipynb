# QQ plot

```python
%matplotlib inline
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt

def qqplot(data):
    n = len(data)
    std = np.std(data)
    mean = np.average(data)
    ids = np.arange(n)
    x = stats.norm.ppf((ids + 0.5) / n)
    y = np.sort(data)
    
    y_norm = np.sort(np.random.normal(loc=mean, scale=std, size=n))    
    
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=50)
    
    plt.subplot(1, 2, 2)
    plt.scatter(x, y, s=20)
    plt.plot(x, y, label='Input Data')
    plt.plot(x, x*std+mean, label='Normal: $\sigma = {:.3f}, \mu = {:.3f}$'.format(std, mean))
    #plt.plot(x, y_norm)
    plt.grid()
    plt.legend()
    
    plt.show()

N = 1000
r = np.random.randn(N)

qqplot(r * 0.2 + 1.5)
qqplot(0.4 * np.exp(-r*0.4))
qqplot(np.random.rand(N))
```
