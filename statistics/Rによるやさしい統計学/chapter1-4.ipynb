```r
# 正規分布の標本平均を10000回取得
sample_mean <- numeric(length=10000)
for(i in 1:10000){
    sample <- rnorm(n=100,mean=50,sd=10)
    sample_mean[i] <- mean(sample)
}
# 標本平均の分布
hist(sample_mean,breaks=seq(40,60,0.2),freq=FALSE)
# 標本誤差は sqrt(母分散/標本数)
curve(dnorm(x,mean=50,sd=1),add=TRUE,col="red")
```

```r
sample_var <- numeric(length=10000)
unbiased_var <- numeric(length=10000)
for(i in 1:10000){
    sample <- rnorm(n=10,mean=50,sd=10)
    sample_var[i] <- mean((sample - mean(sample))^2)
    unbiased_var[i] <- var(sample)
}
mean_sample_var = mean(sample_var)
mean_unbiased_var = mean(unbiased_var)
sprintf("sample_var: %f, unbiased_var: %f", mean_sample_var, mean_unbiased_var)
```

```r
hist(sample_var,breaks=seq(0,400,10))
abline(v=mean_sample_var,col="red",lwd=3)
hist(unbiased_var,breaks=seq(0,400,10))
abline(v=mean_unbiased_var,col="red",lwd=3)
```
