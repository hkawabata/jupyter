```python
from graphviz import Digraph
```

```python
class Node:
    def __init__(self, value, left, right):
        self.value = value
        self.left = left
        self.right = right

def show_tree(nodes):
    dot = Digraph()
    dot.attr('node', shape='circle')
    for n in nodes:
        dot.node(str(n.value), str(n.value))
        if n.left is not None:
            dot.edge(str(n.value), str(n.left))
        if n.right is not None:
            dot.edge(str(n.value), str(n.right))
    return dot

def show_binary_search_tree(nodes):
    dot = Digraph()
    dot.attr('node', shape='record', height='.1')
    for n in nodes:
        dot.node(str(n.value), '<left> |<value>{}|<right> '.format(n.value))
        if n.left is not None:
            #dot.edge(str(n.value), str(n.left))
            dot.edge('{}:left'.format(n.value), '{}:value'.format(n.left))
        if n.right is not None:
            #dot.edge(str(n.value), str(n.right))
            dot.edge('{}:right'.format(n.value), '{}:value'.format(n.right))
    return dot
```

```python
tree = [
    Node(0, 1, 2),
    Node(1, 3, 4),
    Node(2, 5, 6),
    Node(3, 7, 8),
    Node(4, 9, None),
    Node(5, None, None),
    Node(6, None, None),
    Node(7, None, None),
    Node(8, None, None),
    Node(9, None, None)
]

show_tree(tree)
```

```python
binary_search_tree = [
    Node(5, 3, 8),
    Node(1, 0, 2),
    Node(2, None, None),
    Node(3, 1, 4),
    Node(4, None, None),
    Node(6, None, 7),
    Node(7, None, None),
    Node(0, None, None),
    Node(8, 6, 9),
    Node(9, None, None)
]

show_binary_search_tree(binary_search_tree)
```
