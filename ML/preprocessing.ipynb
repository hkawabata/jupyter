```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap

import numpy as np
```

```python
N = 50
x = np.random.rand(N) * 4
y = np.random.rand(N) * 0.6 - 2

x_n, y_n = (x - min(x)) / (max(x) - min(x)), (y - min(y)) / (max(y) - min(y))
x_s, y_s = (x - np.average(x)) / np.std(x), (y - np.average(y)) / np.std(y)

data = np.array([x, y]).T
data_n = np.array([x_n, y_n]).T
data_s = np.array([x_s, y_s]).T

plt.scatter(x, y, s=10, label='Raw')
plt.scatter(x_n, y_n, s=10, label='Normalized')
plt.scatter(x_s, y_s, s=10, label='Standardized')
plt.legend()
plt.grid()
plt.show()
```

```python
N = 360
C = [2, 2]
A = 0.6
B = 3
dR = 0.2
ALPHA = np.pi / 12
ALPHA = np.pi * 5 / 12
th = np.arange(0, N+1) / N * 2*np.pi

plt.axes().set_aspect('equal')
for r in [dR*(i+1) for i in range(7)]:
    x, y = A*r*np.cos(th), B*r*np.sin(th)
    x, y = x*np.cos(ALPHA)-y*np.sin(ALPHA)+C[0], x*np.sin(ALPHA)+y*np.cos(ALPHA)+C[1]
    plt.plot(x, y, c='black')
plt.plot([], [], label='Not Scaled', c='black')
ap = dict(shrink=0, width=0.5, headwidth=5,
        headlength=8, connectionstyle='arc3', facecolor='red', edgecolor='red')
points = [
    [4.6, 2],
    [4, 0.9],
    [4.2, 1.8],
    [3.9, 1.2],
    [3.9, 1.8]
]
for i in range(len(points)-1):
    plt.annotate('', xytext=points[i], xy=points[i+1], arrowprops=ap)

for r in [dR*(i+1) for i in range(7)]:
    x = r * np.cos(th)
    y = r * np.sin(th)
    plt.plot(x, y, c='blue')
plt.plot([], [], label='Scaled', c='blue')
points = [
    [1, 1],
    [0.58, 0.58],
    [0.3, 0.3],
    [0, 0]
]
for i in range(len(points)-1):
    plt.annotate('', xytext=points[i], xy=points[i+1], arrowprops=ap)
plt.grid()
plt.legend()
plt.show()
```

$$
J(\boldsymbol{w}) = f(\boldsymbol{w}) + \lambda \displaystyle \sum_j |w_j|
$$

$$
J(\boldsymbol{w}) = f(\boldsymbol{w}) + \lambda \displaystyle \sum_j w_j^2
$$

$$\|\boldsymbol{w}\|^2 = \displaystyle \sum_j w_j^2$$

$$
\begin{eqnarray}
\min f(\boldsymbol{w}) \\
{\rm subject\ to}\ \|\boldsymbol{w}\|^2 \le C = const.
\end{eqnarray}
$$

$$
\begin{eqnarray}
\min f(\boldsymbol{w}) \\
{\rm subject\ to}\ \displaystyle \sum_j |w_j| \le C = const.
\end{eqnarray}
$$


```python
N = 360
th = np.arange(0, N+1) / N * 2*np.pi

plt.axes().set_aspect('equal')
plt.xlabel('$w_1$')
plt.ylabel('$w_2$')
plt.plot([0, 0], [-1.5, 2], c='gray', linewidth=1)
plt.plot([-1.5, 2], [0, 0], c='gray', linewidth=1)
x = np.cos(th)
y = np.sin(th)
plt.plot(x, y, c='black', label='$R(w) = C$')
x = np.arange(-0.8, 2, 0.1)
y = (x-1)**2 / 3 + 0.8
plt.plot(x, y, c='red', label='$f(w) = const.$')
plt.scatter([0.37], [0.93], c='red')

x = np.arange(-0.999, 1.001, 0.01)
y1 = np.sqrt(1 - x**2)
y2 = -np.sqrt(1 - x**2)
plt.fill_between(x,y1,y2,facecolor='blue',alpha=0.2)

plt.legend()
plt.show()



plt.axes().set_aspect('equal')
plt.xlabel('$w_1$')
plt.ylabel('$w_2$')
plt.plot([0, 0], [-1.5, 2], c='gray', linewidth=1)
plt.plot([-1.5, 2], [0, 0], c='gray', linewidth=1)
x = np.arange(-1, 1.1, 0.1)
y1 = 1 - np.abs(x)
y2 = np.abs(x) - 1
plt.plot(x, y1, c='black')
plt.plot(x, y2, c='black', label='$R(w) = C$')
plt.fill_between(x,y1,y2,facecolor='blue',alpha=0.2)

x = np.arange(-0.8, 2, 0.1)
y = (x-1)**2 / 3 + 0.67
plt.plot(x, y, c='red', label='$f(w) = const.$')
plt.scatter([0], [1], c='red')
plt.legend()
plt.show()
```
