# ヒープ

```python
from graphviz import Digraph

class Heap:
    def __init__(self, array=[]):
        self.data = []
        for v in array:
            self.add(v)

    def add(self, v):
        i = len(self.data)
        self.data.append(v)
        i_parent = int((i-1)/2)
        while i > 0 and self.data[i] < self.data[i_parent]:
            self.data[i], self.data[i_parent] = self.data[i_parent], self.data[i]
            i, i_parent = i_parent, int((i_parent-1)/2)

    def draw_graph(self):
        dot = Digraph()
        dot.attr('node', shape='circle')
        for i in range(len(self.data)):
            dot.node(str(i), str(self.data[i]))
            if i > 0:
                i_parent = int((i-1)/2)
                dot.edge(str(i_parent), str(i))
        return dot
```

```python
h = Heap(['F', 'A', 'J', 'I', 'E', 'B', 'G', 'H', 'D'])
h.data
```

```python
h.draw_graph()
```

```python
h.add('C')
h.draw_graph()
```
