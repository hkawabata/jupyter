# 二分木

```python
class BinaryTree:
    def __init__(self, array):
        self.data = [self.Node(array[0])]
        for v in array[1:]:
            i = 0
            while True:
                if v <= self.data[i].value:
                    if self.data[i].left is None:
                        self.data[i].left = len(self.data)
                        break
                    else:
                        i = self.data[i].left
                elif self.data[i].value < v:
                    if self.data[i].right is None:
                        self.data[i].right = len(self.data)
                        break
                    else:
                        i = self.data[i].right
            self.data.append(self.Node(v))
    
    def search(self, v):
        i = 0
        while True:
            if v < self.data[i].value:
                if self.data[i].left is None:
                    return None
                else:
                    i = self.data[i].left
            elif self.data[i].value < v:
                if self.data[i].right is None:
                    return None
                else:
                    i = self.data[i].right
            else:
                return i
    
    def max_value(self):
        i = 0
        while self.data[i].right is not None:
            i = self.data[i].right
        return self.data[i].value
        
    def min_value(self):
        i = 0
        while self.data[i].left is not None:
            i = self.data[i].left
        return self.data[i].value
    
    class Node:
        def __init__(self, value):
            self.value = value
            self.left = None
            self.right = None
```

```python
tree = BinaryTree(['E', 'A', 'C', 'F', 'H', 'B', 'G', 'J', 'D', 'I'])
[(n.left, n.value, n.right) for n in tree.data]
```

```python
from graphviz import Digraph

dot = Digraph()
dot.attr('node', shape='circle')
for i in range(len(tree.data)):
    value, left, right = tree.data[i].value, tree.data[i].left, tree.data[i].right
    dot.node(str(i), value)
    if left is not None:
        dot.edge(str(i), str(left))
    if right is not None:
        dot.edge(str(i), str(right))
dot
```

```python
for c in ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K']:
    print(tree.search(c))
```
