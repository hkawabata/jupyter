# 8 Queens Problem

- 8x8 のマスにチェスの Queen 8つを並べる
- 8つの Queen は互いに取れない位置に置く

```python
import pandas as pd
```

```python
def n_queens(N):
    x, y = 0, 0
    queens = []
    while 0 <= x and x < N:
        if y < N:
            while y < N:
                ok = True
                for xq, yq in queens:
                    condition = x != xq and y != yq and abs(y-yq) != abs(x-xq)
                    ok = ok and condition
                    if not ok:
                        break
                if ok:
                    break
                y += 1
        else:
            ok = False

        if ok:
            queens.append((x, y))
            x += 1
            y = 0
        else:
            if x == 0:
                break
            x -= 1
            y = queens[-1][1] + 1
            queens = queens[:-1]
    if len(queens) == 0:
        raise Exception('No Answer (N={})'.format(N))

    board = []
    for i in range(N):
        board.append(['.'] * N)
    for xq, yq in queens:
        board[xq][yq] = 'Q'
    return board
```
