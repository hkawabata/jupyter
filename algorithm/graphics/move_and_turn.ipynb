# Move, Turn

```python
# display graph on notebook
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
class MyView:
    def __init__(self, x0=0, y0=0, angle0 = 0):
        self.x = [x0]
        self.y = [y0]
        self.x_now = x0
        self.y_now = y0
        self.angle = angle0
    
    def record(self):
        self.x.append(self.x_now)
        self.y.append(self.y_now)
    
    def draw(self):
        plt.subplot().set_aspect('equal')  # 横長グラフにしないための設定
        plt.plot(self.x, self.y)
        plt.show()
    
    def move(self, d):
        self.x_now += math.cos(self.angle) * d
        self.y_now += math.sin(self.angle) * d
        self.record()
    
    def turn(self, angle):
        self.angle += angle
```

```python
v = MyView()
for i in range(50):
    v.move(10)
    v.turn(math.pi * 0.52)
v.draw()
```

```python
v = MyView()
for i in range(100):
    v.move(10 + 0.1*i)
    v.turn(math.pi * 2.05 / 6)
v.draw()
```

```python
v = MyView()
for i in range(1, 100):
    v.move(10*i)
    v.turn(math.pi / 2.0)
v.draw()

v = MyView()
v.turn(math.pi / 4.0)
for i in range(1, 100):
    v.move(10*i)
    v.turn(math.pi / 2.01)
v.draw()
```
