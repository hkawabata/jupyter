## 散布図

```python
from matplotlib.lines import Line2D

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
plt.scatter(x[:N//2], y[:N//2], marker='x', s=40, color='red', label='foo')
plt.scatter(x[N//2:], y[N//2:], marker='o', s=20, color='blue', label='bar')
plt.legend()
plt.show()

x, y = np.array([0]), np.array([0])
cnt = 0
for k, v in Line2D.markers.items():
    if cnt%10 == 0:
        x = 0
        y += 1
    x += 1
    plt.scatter(x, y, marker=k, label='{} {}'.format(k, v))
    cnt += 1
plt.legend(ncol=3, bbox_to_anchor=(1.05, 0.93, 0.5, .100))
plt.show()
```

## 折れ線グラフ

```python
x = np.arange(0, 10+1, 1)
y1 = x**2
y2 = x**1.5
plt.plot(x, y1, color='red', label='foo')
plt.plot(x, y2, linewidth=3.0, color='blue', label='bar')
plt.legend()
plt.show()

# linestyle
x = np.arange(0, 10.0+1, 1.0)
y = x * 0.2
linestyles = [
    'solid', 'dashed', 'dotted', 'dashdot',
    (0, (1, 10)),
    (0, (5, 10)),
    (0, (5, 1)),
    (0, (3, 10, 1, 10)),
    (0, (3, 1, 1, 1)),
    (0, (3, 5, 1, 5, 1, 5)),
    (0, (3, 10, 1, 10, 1, 10)),
    (0, (3, 1, 1, 1, 1, 1))
]
for style in linestyles:
    plt.plot(x, y, linestyle=style, label='{}'.format(style))
    y += 1.0
plt.legend(ncol=1, bbox_to_anchor=(1.05, 0.93, 0.5, .100))
plt.show()
```

## 等高線

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
x, y = np.meshgrid(x, y)
z = 2*np.exp(-((x+3)**2+y**2)/2) - np.exp(-((x-2)**2+(y-2)**2)/2) - 2*np.exp(-((x-2)**2+(y+2)**2)/2)

plt.contour(x, y, z, 20)
plt.show()

cs = plt.contour(x, y, z, 20)
plt.clabel(cs, inline=1, fontsize=10)
plt.show()

plt.contourf(x, y, z, 20, cmap='gray')
plt.show()
```

```python
x = np.arange(-1, 11, 0.02)

def f1(x):
    return -x / 4.0 + 5.5
def f2(x):
    return -2 * x + 9
def f3(x):
    return x / 2.0 + 2

y1 = f1(x)
y2 = f2(x)
y3 = f3(x)
ymax = pylab.minimum(y1, y2)
ymin = y3

plt.xlim([0,8])
plt.ylim([0,8])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('maximize $z = x + 2y$')

plt.plot(x, -x/2.0+6, color='blue', linewidth=2.0, label='optimal solution')
plt.plot(x, y1, color='black', linewidth=0.5)
plt.plot(x, y2, color='black', linewidth=0.5)
plt.plot(x, y3, color='black', linewidth=0.5)
plt.fill_between(x, ymin, ymax, where=ymax>ymin, facecolor='gray',alpha=0.4)

plt.annotate('', xytext=[0,0], xy=[0,2],
    arrowprops=dict(shrink=0, width=1, headwidth=8, 
        headlength=10, connectionstyle='arc3', facecolor='green', edgecolor='green'))
plt.annotate('', xytext=[0,2], xy=[14.0/5,17.0/5],
    arrowprops=dict(shrink=0, width=1, headwidth=8, 
        headlength=10, connectionstyle='arc3', facecolor='green', edgecolor='green'))
plt.annotate('', xytext=[14.0/5,17.0/5], xy=[2,5],
    arrowprops=dict(shrink=0, width=1, headwidth=8, 
        headlength=10, connectionstyle='arc3', facecolor='green', edgecolor='green'))

plt.legend()
plt.grid()

plt.show()
```
