```python
# display graph on notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pylab
import math
```

```python
def f1(x):
    return - x / 3.0 + 6000
def f2(x):
    return -x * 2 / 3.0 + 7000
def f3(x):
    return -3 * x + 21000

x = np.arange(-500, 10001, 50)
y1 = f1(x)
y2 = f2(x)
y3 = f3(x)
ymax = pylab.minimum(pylab.minimum(y1, y2), y3)

plt.xlim([0,8000])
plt.ylim([0,16000])
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

plt.plot(x, -2*x+15000, label='$2x_1 + x_2 = 15000$')
plt.plot(x, y1, color='black', linewidth=0.5)
plt.plot(x, y2, color='black', linewidth=0.5)
plt.plot(x, y3, color='black', linewidth=0.5)
plt.fill_between(x, 0, ymax, where=ymax>0, facecolor='gray',alpha=0.5)
plt.legend()

plt.show()
```
